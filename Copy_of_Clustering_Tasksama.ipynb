## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
# remove the outliers
# we chose only numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1

# calcute based on IQR
df = df[~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Select relevant numerical features for scaling
numerical_features = ['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude', 'Latitude', 'Number_of_Vehicles', 'Number_of_Casualties']
```

```python
# Scale numerical features
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)  # Choose number of components
df_pca = pca.fit_transform(df[numerical_features])
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
# matrix empty
inertia=[]
for i in range(1,11):
  kmeans =KMeans(n_clusters=i, random_state=42)
  kmeans.fit(numeric_cols[['Longitude','Latitude']])
  inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),inertia,marker='o')
#plt.titla('Elbow Method')
plt.title("Elbow Method")
plt.xlabel("number of K")
plt.ylabel("Inertia")
plt.show()
```

```python
#taining the model
optimal_k=3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
numeric_cols['cluster'] = kmeans.fit_predict(numeric_cols[['Longitude', 'Latitude']])
```
